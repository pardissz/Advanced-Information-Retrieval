<p>
در این بخش باید برای سامانه یک
Positional Index

<br>
با توجه به مواردی که در بخش بعد می‌آید و نیاز به جست‌وجو‌ی مجزا و با امتیازدهی متفاوت بر روی بخش‌های مختلف سند مثل عنوان یا چکیده آن، در این قسمت باید نمایه‌ی مناسب برای امکان جست‌وجو‌ در بخش‌های مختلف را پیاده‌سازی کنید.
با استفاده از نمایه‌ی ساخته‌شده باید بتوان شماره‌ی تمامی اسنادی که یک کلمه در آن آمده است و همچنین همه‌ی جایگاه‌های این کلمه در هر بخش از هر سند را پیدا کرد.

برای این بخش می‌توانید از نمایه‌ای که در تمرین اول زدید استفاده کنید.
</p>
</div>

```python
def construct_positional_indexes(df: pd.DataFrame) -> dict:
    """
    Constructs positional indexes for the processed data by inserting words into a trie and creating positional indexes and posting lists.

    Parameters
    ----------
    df : pd.DataFrame
        The processed data.

    Returns
    -------
    dict
        The constructed positional indexes.
    """

    positional_index = defaultdict(lambda: {'total_occurrences': 0, 'documents': defaultdict(
        lambda: {'title': {'doc_occurrences': 0, 'positions': []},
                 'abstract': {'doc_occurrences': 0, 'positions': []}})})
    for index, row in df.iterrows():
        paper_id = row['paperId']
        title_tokens = nltk.word_tokenize(row['preprocessed_title'])
        abstract_tokens = nltk.word_tokenize(row['preprocessed_abstract'])
        for position, token in enumerate(title_tokens):
            positional_index[token]['total_occurrences'] += 1
            positional_index[token]['documents'][paper_id]['title']['doc_occurrences'] += 1
            positional_index[token]['documents'][paper_id]['title']['positions'].append(position)
        for position, token in enumerate(abstract_tokens):
            positional_index[token]['total_occurrences'] += 1
            positional_index[token]['documents'][paper_id]['abstract']['doc_occurrences'] += 1
            positional_index[token]['documents'][paper_id]['abstract']['positions'].append(position)

    return positional_index


positional_index = construct_positional_indexes(df)
```

<h2>2-1.
ساخت ماتریس TF-IDF
</h2>
<p>
در این مرحله دو تابع
<code dir="ltr">get_tf(token, doc_id)</code>
و
<code dir="ltr">get_idf(token)</code>
را پیاده‌سازی کنید که تابع اول مقدار
tf
توکن ورودی را در شناسه‌ی موردنظر
و تابع دوم مقدار
idf
توکن ورودی را برروی نمایه‌ی کنونی حساب می‌کند و آن را به عنوان خروجی برمی‌گرداند.
</p>
</div>

$$tf_{t, d} = \text{Numbers of } t \text{ in the title of document } d + \text{Numbers of } t \text{ in the abstract of document } d$$
$$idf_t = \log\Biggl(\frac{\text{Number of documents}}{\text{Number of documents that contains } t + 1}\Biggr)$$

```python
def get_tf(token: str, positional_index: dict, doc_id: str) -> int:
    """
    Retrieves the term frequency (TF) of a given token within a specific document.

    Parameters
    ----------
    token : str
        The token for which to retrieve the term frequency.
    positional_index : dict
        A positional index that maps tokens to their positions in documents.
    doc_id : str
        The unique identifier of the document in which to calculate the TF.

    Returns
    -------
    int
        The term frequency (TF) of the given token within the specified document.
    """
    title_positions = positional_index[token]['documents'][doc_id]['title']['positions']
    abstract_positions = positional_index[token]['documents'][doc_id]['abstract']['positions']
    tf = len(title_positions) + len(abstract_positions)
    return tf


print(get_tf('measure', positional_index, 'c8d96c09593bd7244449b24a479bff8284c38f79'))
```

```python
import math


def get_idf(token: str, positional_index: dict, total_docs: int) -> float:
    """
    Retrieves the inverse document frequency (IDF) of a given token.

    Parameters
    ----------
    token : str
        The token to retrieve the IDF for.
    positional_index : dict
        A positional index that maps tokens to their positions in documents.
    total_docs : int
        The total number of documents in the collection.

    Returns
    -------
    float
        The IDF of the given token.
    """

    num_docs_with_token = len(positional_index[token]['documents'])
    idf = math.log(total_docs / (num_docs_with_token + 1))
    return idf


print(get_idf('tesla', positional_index, len(df)))
print(get_idf('report', positional_index, len(df)))

```

<p>
در مرحله‌ی نهایی تابع

را پیاده‌سازی کنید که با ورودی گرفتن
corpus

مقادیر
tf-idf
را برای هر

به دست می‌آورد و به شکل یک ماتریس به عنوان خروجی برمی‌گرداند.

```python

def generate_tfidf_list(corpus: pd.DataFrame, positional_index: dict) -> list[dict[str, float]]:
    """
    Generates a list of dictionaries representing documents with associated TF-IDF scores.

    Parameters
    ----------
    corpus : pd.DataFrame
        The corpus containing the processed data.
    positional_index : dict
        The positional index containing the term frequencies and document frequencies.

    Returns
    -------
    list[dict[str, float]]
        The list of dictionaries representing documents with associated TF-IDF scores.
    """
    tfidf_list = []
    total_docs = len(corpus)
    for index, row in corpus.iterrows():
        paper_id = row['paperId']
        tfidf_dict = {'paperId': paper_id}
        title_tokens = nltk.word_tokenize(row['preprocessed_title'])
        abstract_tokens = nltk.word_tokenize(row['preprocessed_abstract'])
        for token in set(title_tokens + abstract_tokens):
            tf = get_tf(token, positional_index, paper_id)
            idf = get_idf(token, positional_index, total_docs)
            tfidf = tf * idf
            tfidf_dict[token] = tfidf

        tfidf_list.append(tfidf_dict)

    return tfidf_list


tfidf = generate_tfidf_list(df, positional_index)
```

```python

def show_document_vector(vector: list[dict[str, float]], doc_index: int) -> None:
    """
    Prints the non-zero weights and corresponding terms for a document represented as a vector.

    Parameters
    ----------
    vector : list of dict of str:float
        The list of document vectors, where each document vector is a dictionary with term weights.
    doc_id : int
        The document index for which to display the vector.

    Returns
    -------
    None
    """
    vector_list = [(term, weight) for term, weight in vector[doc_index].items() if term != 'paperId' and weight > 0]
    vector_list.sort(key=lambda x: x[1], reverse=True)
    length = math.sqrt(sum(weight ** 2 for _, weight in vector_list))
    normalized = {term: tfidf / length for term, tfidf in vector_list}
    for term, tfidf in vector_list:
        print(f'{term.ljust(16)}\t{tfidf}\t(normalized: {normalized[term]})')


show_document_vector(tfidf, 0)
```

</p>
</div>
<div>
<p>

$$ RSV_d = \sum_{t \in q} idf(t)\times\frac{(k_1 + 1)tf(t, d)}{k_1((1 - b) + b\frac{dl(d)}{avg(dl)}) + tf(t, d)} $$
</p>
</div>
<div dir='rtl'>
<p>
برای محاسبه این دو معیار، از توابعی که در بخش قبل پیاده‌سازی کردید استفاده کنید.

در تابع search که مربوط به جست و جوی پرسمان کلی است، به عنوان ورودی پارامتر پرسمان (query)، روش محاسبه امتیاز (method)، تعداد اسنادی که باید برگردانده شود (max-result-count) را ورودی می گیرید. ورودی وزن
(weight)
نشان می‌دهد که امتیاز نهایی چه وزنی از امتیاز عنوان و چکیده دارد. به زبان دیگر:
</p>
</div>
<div>
<p>

$$ final\_score = weight \times title\_score + (1 - weight) \times abstract\_score $$
</p>
</div>
<div dir='rtl'>
<p>

**توجّه کنید** که تابع نوشته شده، صرفاً یک prototype است و تا وقتی که نیازمندی‌های گفته شده را پیاده کنید، تغییر تابع مشکلی ندارد.
</p>
</div>

```python
def search(query: str, max_result_count: int, method: str = 'ltc-lnc-tat', weight: float = 0.7):
    """
    Finds relevant documents to query

    Parameters
    ---------------------------------------------------------------------------------------------------
    max_result_count: Return top 'max_result_count' docs which have the highest scores.
                        notice that if max_result_count = -1, then you have to return all docs

    method: 'ltc-lnc-tat' or 'ltc-lnc-dat' or 'okapi25'

    Returns
    ----------------------------------------------------------------------------------------------------
    list
    Retreived documents with snippet
    """
    preprocessor = Preprocessor()
    query=preprocessor.preprocess(query)
    if method == 'ltc-lnc-tat':
        query_tokens = nltk.word_tokenize(query)
        doc_scores = defaultdict(float)
        query_length = math.sqrt(len(query_tokens))
        for token in query_tokens:
            postings_list = positional_index[token]['documents']
            idf = get_idf(token, positional_index, len(df))
            query_tf = query_tokens.count(token)
            query_weight = (1 + math.log(query_tf)) / query_length
            for doc_id, doc_data in postings_list.items():
                title_tf = doc_data['title']['doc_occurrences']
                abstract_tf = doc_data['abstract']['doc_occurrences']
                title_length = math.sqrt(title_tf)
                abstract_length = math.sqrt(abstract_tf)
                title_score = ((1 + math.log(title_tf + 1)) * idf) / title_length if title_length > 0 else 0
                abstract_score = ((1 + math.log(abstract_tf + 1)) * idf) / abstract_length if abstract_length > 0 else 0
                doc_scores[doc_id] += query_weight * (weight * title_score + (1 - weight) * abstract_score)
        sorted_docs = sorted(doc_scores.items(), key=lambda x: x[1], reverse=True)
        if max_result_count != -1:
            sorted_docs = sorted_docs[:max_result_count]

        return sorted_docs
    elif method == 'ltc-lnc-dat':
        query_tokens = nltk.word_tokenize(query)
        doc_scores = defaultdict(float)
        query_length = math.sqrt(len(query_tokens))
        for index, row in df.iterrows():
            doc_id = row['paperId']
            title_tokens = nltk.word_tokenize(row['preprocessed_title'])
            abstract_tokens = nltk.word_tokenize(row['preprocessed_abstract'])
            for token in query_tokens:
                if token not in title_tokens + abstract_tokens:
                    continue
                postings_list = positional_index[token]['documents']
                idf = get_idf(token, positional_index, len(df))
                query_tf = query_tokens.count(token)
                query_weight = (1 + math.log(query_tf)) / query_length
                title_tf = postings_list[doc_id]['title']['doc_occurrences']
                abstract_tf = postings_list[doc_id]['abstract']['doc_occurrences']
                title_length = math.sqrt(title_tf)
                abstract_length = math.sqrt(abstract_tf)
                title_score = ((1 + math.log(title_tf + 1)) * idf) / title_length if title_length > 0 else 0
                abstract_score = ((1 + math.log(abstract_tf + 1)) * idf) / abstract_length if abstract_length > 0 else 0
                doc_scores[doc_id] += query_weight * (weight * title_score + (1 - weight) * abstract_score)
        sorted_docs = sorted(doc_scores.items(), key=lambda x: x[1], reverse=True)
        if max_result_count != -1:
            sorted_docs = sorted_docs[:max_result_count]

        return sorted_docs
    elif method == 'okapi25':
        k1=1.2
        b= 0.75
        query_tokens = nltk.word_tokenize(query)
        doc_scores = defaultdict(float)
        avg_title_length = sum(len(nltk.word_tokenize(row['preprocessed_title'])) for _, row in df.iterrows()) / len(df)
        avg_abstract_length = sum(
            len(nltk.word_tokenize(row['preprocessed_abstract'])) for _, row in df.iterrows()) / len(df)
        for index, row in df.iterrows():
            doc_id = row['paperId']
            title_tokens = nltk.word_tokenize(row['preprocessed_title'])
            abstract_tokens = nltk.word_tokenize(row['preprocessed_abstract'])
            title_length = len(title_tokens)
            abstract_length = len(abstract_tokens)
            for token in query_tokens:
                if token not in title_tokens + abstract_tokens:
                    continue
                postings_list = positional_index[token]['documents']
                idf = get_idf(token, positional_index, len(df))
                title_tf = postings_list[doc_id]['title']['doc_occurrences']
                abstract_tf = postings_list[doc_id]['abstract']['doc_occurrences']
                title_score = idf * ((k1 + 1) * title_tf) / (k1 * (
                            (1 - b) + b * (title_length / avg_title_length)) + title_tf) if title_length > 0 else 0
                abstract_score = idf * ((k1 + 1) * abstract_tf) / (k1 * ((1 - b) + b * (
                            abstract_length / avg_abstract_length)) + abstract_tf) if abstract_length > 0 else 0
                doc_scores[doc_id] += weight * title_score + (1 - weight) * abstract_score
        sorted_docs = sorted(doc_scores.items(), key=lambda x: x[1], reverse=True)
        if max_result_count != -1:
            sorted_docs = sorted_docs[:max_result_count]

        return sorted_docs
a=search("application compromise evolution",3,'okapi25',0.7)
print(a)
```

To check if I have dont it correctly I print the metric for each query as well and then print the average for the queries. If you want you can comment the print statements in code below

```python
class Evaluation:
    def __init__(self, actual, predicts):
        self.actual = actual
        self.predicts = [list(map(lambda x: x[0], predict)) for predict in predicts]
    def evaluate(self, require_scores):
        """
        Prints require scores for actual and predicts array

        Parameters
        ----------
        require_scores : List[str]
            Scores required to calculated

        Returns
        -------
        """
        scores = {}
        for score in require_scores:
            if score == 'precision':
                scores['precision'] = self.precision()
            elif score == 'recall':
                scores['recall'] = self.recall()
            elif score == 'f1':
                scores['f1'] = self.f1()
            elif score == 'map':
                scores['map'] = self.map()
            elif score == 'ndcg':
                scores['ndcg'] = self.ndcg()
            elif score == 'mrr':
                scores['mrr'] = self.mrr()
        return scores

    def precision(self, print_results=True):
        """
        Calculates the precision of the predicted results

        Parameters
        ----------
        print_results : bool, optional
            Whether to print the results, by default True

        Returns
        -------
        float
            The precision of the predicted results
        """
        precisions = []
        for a, p in zip(self.actual, self.predicts):
            binary = [1 if i in a else 0 for i in p]
            tp = sum(binary)
            fp = len(binary) - tp
            precision = tp / (tp + fp) if tp + fp > 0 else 0
            precisions.append(precision)
        if print_results:
            print("precisions for each query is: " + str(precisions))
        return np.mean(precisions)

    def recall(self, print_results=True):
        """
        Calculates the recall of the predicted results

        Parameters
        ----------
        print_results : bool, optional
            Whether to print the results, by default True

        Returns
        -------
        float
            The recall of the predicted results
        """
        recalls = []
        for a, p in zip(self.actual, self.predicts):
            binary = [1 if i in a else 0 for i in p]
            tp = sum(binary)
            fn = len(a) - tp
            recall = tp / (tp + fn) if tp + fn > 0 else 0
            recalls.append(recall)
        if print_results:
            print("recalls for each query is: " + str(recalls))
        return np.mean(recalls)

    def f1(self):
        """
        Calculates the F1 score of the predicted results

        Returns
        -------
        float
            The F1 score of the predicted results
        """
        precision = self.precision(print_results=False)
        recall = self.recall(print_results=False)
        f1_score = 2 * (precision * recall) / (precision + recall)

        return f1_score
    def map(self):
        """
        Calculates the Mean Average Precision of the predicted results

        Returns
        -------
        float
            The Mean Average Precision of the predicted results
        """
        maps = []
        for a, p in zip(self.actual, self.predicts):
            ap = 0.0
            num_rel = 0
            for i, doc_id in enumerate(p, start=1):
                if doc_id in a:
                    num_rel += 1
                    ap += num_rel / i
            maps.append(ap / num_rel if num_rel else 0)
        return np.mean(maps)

    def ndcg(self):
        """
        Calculates the Normalized Discounted Cumulative Gain (NDCG) of the predicted results

        Returns
        -------
        float
            The NDCG of the predicted results
        """
        ndcgs = []
        for a, p in zip(self.actual, self.predicts):
            dcg = sum([int(p[i] in a)/np.log2(i+2) for i in range(len(p))])
            idcg = sum([1/np.log2(i+2) for i in range(len(a))])
            ndcg = dcg/idcg
            ndcgs.append(ndcg)
        print("ndcgs for each query is: " + str(ndcgs))
        ndcg_score = np.mean(ndcgs)

        return ndcg_score

    def mrr(self):
        """
        Calculates the Mean Reciprocal Rank of the predicted results

        Returns
        -------
        float
            The MRR of the predicted results
        """
        rranks = []
        for a, p in zip(self.actual, self.predicts):
            rranks.append(1 / (p.index(next(x for x in p if x in a)) + 1))
        print("rranks for each query is: " + str(rranks))

        mrr_score = np.mean(rranks)

        return mrr_score
```
