```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import spacy
import re
nltk.download('punkt')
nltk.download('stopwords')
class Preprocessor:
    def __init__(self):

        self.stopwords = set(stopwords.words('english'))
        self.nlp = spacy.load('en_core_web_sm')

    def preprocess(self, text):
        text = self.normalize(text)
        text = self.remove_links(text)
        text = self.remove_punctuations(text)
        words = self.word_tokenize(text)
        words = self.remove_stopwords(words)
        return ' '.join(words)

    def normalize(self, text):
        text = text.lower()
        doc = self.nlp(text)
        text = ' '.join([token.lemma_ for token in doc])

        return text

    def remove_links(self, text):
        patterns = [r'http\S+', r'www\S+', r'https\S+']
        for pattern in patterns:
            text = re.sub(pattern, '', text)
        return text

    def remove_punctuations(self, text):
        return re.sub(r'[^\w\s]', '', text)

    def word_tokenize(self, text):
        return word_tokenize(text)

    def remove_stopwords(self, words):
        return [word for word in words if word not in self.stopwords]
```

```python
preprocessor = Preprocessor()
df['preprocessed_abstract'] = df['abstract'].apply(preprocessor.preprocess)
df.to_csv('processed_data.csv', index=False)
```

```python
from collections import defaultdict, deque
import hashlib
class DocumentBatch:
    def __init__(self, docs, start_id=0):
        self.documents = docs
        self.index = defaultdict(set)
        self.start_id = start_id

    def sort_based_index_construction(self):
        for doc_id, doc in enumerate(self.documents, start=self.start_id):
            preprocessor = Preprocessor()
            preprocessed_text = preprocessor.preprocess(doc)
            tokens = preprocessor.word_tokenize(preprocessed_text)
            for token in tokens:
                self.index[token].add(hashlib.sha256(str(doc_id).encode()).hexdigest())

class FastSearchEngine:
    def __init__(self):
        self.main_index = defaultdict(set)
        self.daily_indices = deque()

    def add_batch(self, batch: DocumentBatch):
        batch.sort_based_index_construction()
        self.daily_indices.append(batch.index)

    def end_of_day_logarithmic_merge(self):
        while len(self.daily_indices) > 1:
            index1 = self.daily_indices.popleft()
            index2 = self.daily_indices.popleft()
            merged_index = defaultdict(set)
            for key in set(list(index1.keys()) + list(index2.keys())):
                merged_index[key] = index1[key].union(index2[key])
            self.daily_indices.append(merged_index)
        if self.daily_indices:
            daily_index = self.daily_indices.popleft()
            for key in set(list(self.main_index.keys()) + list(daily_index.keys())):
                self.main_index[key] = self.main_index[key].union(daily_index[key])

server_a_docs = ["sample_1", "sample_2 is good", "sample_3 I love it", "sample_4 good", "sample_5 good", "sample_6", "sample_7", "sample_8", "sample_9", "sample_10"]
server_b_docs = ["sample_11", "sample_2 is nice", "sample_3", "sample_4 love", "sample_15", "sample_16", "sample_17", "sample_18", "sample_19", "sample_20"]
search_engine = FastSearchEngine()
for i in range(5):
    server_a_batch = DocumentBatch(server_a_docs[i*2:i*2+2], start_id=i*2)
    server_b_batch = DocumentBatch(server_b_docs[i*2:i*2+2], start_id=len(server_a_docs)+i*2)

    search_engine.add_batch(server_a_batch)
    search_engine.add_batch(server_b_batch)

    search_engine.end_of_day_logarithmic_merge()

for key, value in search_engine.main_index.items():
    print(f"{key} : {value}")
```

<div dir='rtl'>
<h2>پاسخ سوال بالا</h2>
پاسخ‌ خود را در این سلول بنویسید.


Minimum Edit Distance: این اندازه گیری عدم مشابهت بین دو رشته است که با کمک محاسبه حداقل تعداد عملیات (درج، حذف و جایگزینی) مورد نیاز برای تبدیل یک رشته به رشته دیگر محاسبه می شود. با یافتن "نزدیک ترین" کلمات در فرهنگ لغت به یک کلمه غلط املایی را پیدا می کنیم.


 N-gram: تجزیه متن به تکه های متوالی است  و شناسایی توالی های رایج کاراکترها یا کلمات کمک کننده است.


Phonetic Algorithms: این الگوریتم ها مانند Soundex یا  Metaphone، کلمات را بر اساس تلفظ آنها رمزگذاری می کنند.


Probabilistic Models: این مدل‌ها بر اساس عواملی مانند فراوانی کلمه در زبان و احتمال اشتباهات املایی خاص، احتمال را برای کلمه مورد نظر محاسبه می‌کنند. تصحیح با بیشترین احتمال انتخاب شده.
یادگیری ماشینی: الگوریتم‌های یادگیری ماشینی را می‌توان بر روی مجموعه داده‌های بزرگی از خطاها و اصلاحات رایج آموزش داد تا املای صحیح کلمات غلط املایی را پیش‌بینی کند. مدل‌های یادگیری عمیق حتی می‌توانند الگوها و وابستگی‌های پیچیده را در داده‌ها ثبت کنند.

Online spellchecking: با استفاده از جفت‌های تصحیح املایی، یک مدل تبدیل مارکوف n-gram آموزش داده می شود که رفتار املایی را به شیوه‌ای بدون نظارت unsupervised ثبت ‌کند. برای یافتن بهترین پیشنهادات پرسمان تصحیح شده، الگوریتم جستجوی A* با هیوریستیک های مختلف برای گسترش پویا فضای جستجو استفاده می شود.

Jaccard score:ضریب شباهت جاکارد، برای مقایسه شباهت و تفاوت داده ها استفاده می‌شود.
ضریب جاکارد از 0 تا 1 متغیر است، که در آن 0 به این معنی است که مجموعه ها هیچ همپوشانی ندارند و 1 به معنای یکسان بودن مجموعه ها است.

موتورهای جستجو پرسمان های اصلاح شده را معمولا به شکل زیر به کاربر نمایش می دهند:

پرسمان از UI که در آن کاربر عبارت جستجو را وارد می کند، گرفته می شود.

پرسمان ممکن است با استفاده از تکنیک هایی مانند بررسی spelling یا query expansions پردازش می شود.

سپس پرسمان تجزیه شده و به index server ارسال می شود.

در ادامه اندازه‌ی هر کدام از سه فایل بالا را با استفاده از یک تابع به دست آورده و چاپ کنید.  
همچنین باید تابع‌هایی برای decompress کردن نمایه‌های فشرده‌شده پیاده‌سازی کنید.

**نکته‌ی ۱:** تمامی نمایه‌ها را نیز در کوئرا ارسال کنید. اگر حجم‌شان بیش‌تر از محدودیت کوئرا است، آن‌ها را در یک مکان دیگر آپلود کرده و لینک آن را در این فایل قرار دهید.  
**نکته‌ی ۲:** توابع زیر صرفاً پیشنهادی هستند و هر گونه تغییر تا زمانی که کاربردهای مورد نظر پیاده شود، آزاد است.

```python
import os
import pickle

def gamma_encode(number):
    binary = bin(number)[2:]
    length = len(binary)
    offset_binary = binary[1:]
    a=  '1' * (length - 1) + '0' + offset_binary
    return a


def variable_byte_encode(number):
    bytes_list = []
    while True:
        bytes_list.insert(0, number % 128)
        if number < 128:
            break
        number = number // 128
    bytes_list[-1] += 128
    return ''.join(format(byte, '08b') for byte in bytes_list)

def gamma_decode(code):
    length = code.find('0')
    if length == -1 or length == len(code) - 1:
        return 1
    binary = '1' + code[length+1: 2*length+1]
    if binary == '':
        return None
    return int(binary, 2)

def variable_byte_decode(binary_list):
    binary_string = ''.join([binary_list[i] for i in range(len(binary_list)) if (i) % 8 != 0])
    decimal_number = int(binary_string, 2)
    return decimal_number

def save_index(index, filename):
    with open(filename, 'wb') as f:
        pickle.dump(index, f)
def save_uncompressed_index(index, filename):
    binary_index = {k: {bin(i)[2:].zfill(20) for i in v} for k, v in index.items()}
    with open(filename, 'wb') as f:
        pickle.dump(binary_index, f)

def load_index(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)

def get_size(filename):
    return os.path.getsize(filename)

def save_compressed_index(index, filename, encode_method):
    compressed_index = {}
    for k, v in index.items():
        sorted_ids = sorted(v)
        gaps = [sorted_ids[0]] + [sorted_ids[i] - sorted_ids[i-1] for i in range(1, len(sorted_ids))]
        compressed_index[k] = [encode_method(i) for i in gaps]
    save_index(compressed_index, filename)
def uncompressed_decode(binary_string):
    decimal_number = int(binary_string, 2)
    return decimal_number

def load_compressed_index(filename, decode_method):
    compressed_index = load_index(filename)
    decompressed_index = {}
    for k, v in compressed_index.items():
        gaps = [decode_method(num) for num in v]
        doc_ids = [gaps[0]]
        for gap in gaps[1:]:
            doc_ids.append(doc_ids[-1] + gap)
        decompressed_index[k] = doc_ids
    return decompressed_index
def load_uncompressed_index(filename, decode_method):
    compressed_index = load_index(filename)
    decompressed_index = {}
    for k, v in compressed_index.items():
        doc_ids = [decode_method(num) for num in v]
        decompressed_index[k] = doc_ids
    return decompressed_index

a=search_engine.main_index
print(a)
save_uncompressed_index(search_engine.main_index, 'uncompressed_index.pkl')

save_compressed_index(search_engine.main_index, 'gamma_compressed_index.pkl', gamma_encode)

save_compressed_index(search_engine.main_index, 'vb_compressed_index.pkl', variable_byte_encode)

print("Size of uncompressed index (20 bits): ", get_size('uncompressed_index.pkl'))
print("Size of variable byte compressed index: ", get_size('vb_compressed_index.pkl'))
print("Size of gamma compressed index: ", get_size('gamma_compressed_index.pkl'))

gamma_decompressed_index = load_compressed_index('gamma_compressed_index.pkl', gamma_decode)

vb_decompressed_index = load_compressed_index('vb_compressed_index.pkl', variable_byte_decode)
uncompressed_index = load_uncompressed_index('uncompressed_index.pkl', uncompressed_decode)

if gamma_decompressed_index == vb_decompressed_index:
    print("Working fine")
else:
    print("Wrong")
print("decompressed index (20 bits) after decoding: ",uncompressed_index)
sorted_uncompressed_index = {k: sorted(v) for k, v in uncompressed_index.items()}
print("decompressed index (20 bits) after decoding and sorting: ",sorted_uncompressed_index)
print("gamma_decompressed_index: ",gamma_decompressed_index)
print("vb_decompressed_index: ",vb_decompressed_index)
```
